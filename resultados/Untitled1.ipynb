```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df_mutacao_correto = pd.read_csv('eda.csv',index_col=0)
df_a_correto = df_mutacao_correto.loc[df_mutacao_correto['makespan'].isin(df_mutacao_correto.groupby('instance')['makespan'].min().values)]
df_a_correto.sort_values('instance',inplace=True, ascending=True)
```

```python
df_to_compare = pd.DataFrame(columns=['Algorithm', 'Instance', 'Passamos', 'Quanto falta','Makespan do algoritmo', 'Nosso makespan'])
df_to_compare['Instance'] = ['u_c_hihi.0', 'u_c_hilo.0', 'u_c_lohi.0', 'u_c_lolo.0','u_i_hihi.0','u_i_hilo.0',
    'u_i_lohi.0','u_i_lolo.0','u_s_hihi.0', 'u_s_hilo.0', 'u_s_lohi.0', 'u_s_lolo.0']
df_to_compare['Algorithm'] = 'pCHC - average'
df_to_compare['Nosso makespan'] = list(df_a_correto.drop_duplicates(subset=['makespan'])['makespan'])
df_to_compare['Makespan do algoritmo'] = [7481195,153924,243446.3,5181.6, 2956906,73847.1,102677.3,2557.2,4239146,96750.3,123989.4,3472.2]
df_to_compare.loc[df_to_compare['Makespan do algoritmo'] > df_to_compare['Nosso makespan'], 'Passamos'] = 'True' 
df_to_compare.loc[df_to_compare['Makespan do algoritmo'] < df_to_compare['Nosso makespan'], 'Passamos'] = 'False' 
df_to_compare['Quanto falta'] =  df_to_compare['Nosso makespan'] - df_to_compare['Makespan do algoritmo']
```
